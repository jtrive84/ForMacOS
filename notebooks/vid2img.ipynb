```python
import sys
import os
import os.path
import numpy as np
from scipy import ndimage
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

np.set_printoptions(suppress=True, precision=5)
np.set_printoptions(linewidth=1000)
np.core.arrayprint._line_width = 1000


# Import functions from ps03.py.
sys.path.append("G:/Repos/GT/CS6476/ps03")
import ps3
import experiment


tmpl_path = "G:/Repos/GT/CS6476/ps03/input_images/template.jpg"
y_path0 = "G:/Repos/GT/CS6476/ps03/input_videos/ps3-4-a.mp4"
y_path1 = "G:/Repos/GT/CS6476/ps03/input_videos/ps3-4-b.mp4"
y_path2 = "G:/Repos/GT/CS6476/ps03/input_videos/ps3-4-c.mp4"
y_path3 = "G:/Repos/GT/CS6476/ps03/input_videos/ps3-4-d.mp4"

tmpl = cv2.imread(tmpl_path)


outvids0 = "G:/Repos/GT/CS6476/ps03/output/outvids0"
outvids1 = "G:/Repos/GT/CS6476/ps03/output/outvids1"
outvids2 = "G:/Repos/GT/CS6476/ps03/output/outvids2"
outvids3 = "G:/Repos/GT/CS6476/ps03/output/outvids3"

outimgs0 = "G:/Repos/GT/CS6476/ps03/output/outimgs0"
outimgs1 = "G:/Repos/GT/CS6476/ps03/output/outimgs1"
outimgs2 = "G:/Repos/GT/CS6476/ps03/output/outimgs2"
outimgs3 = "G:/Repos/GT/CS6476/ps03/output/outimgs3"

vidpaths = [y_path0, y_path1, y_path2, y_path3]
viddirs = [outvids0, outvids1, outvids2, outvids3]
imgdirs = [outimgs0, outimgs1, outimgs2, outimgs3]



for viddir in viddirs:
    vid_paths = [os.path.join(viddir, ii) for ii in os.listdir(viddir)]
    dev_null = [os.remove(ii) for ii in vid_paths]
    
for imgdir in imgdirs:
    img_paths = [os.path.join(imgdir, ii) for ii in os.listdir(imgdir)]
    dev_null = [os.remove(ii) for ii in img_paths]

for vidpath, viddir, imgdir in zip(vidpaths, viddirs, imgdirs):
    cap = cv2.VideoCapture(vidpath)
    ii = 0
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret==False:
            break
        frame_path = "{}/{}.png".format(viddir, str(ii).zfill(4))
        cv2.imwrite(frame_path, frame)
        ii+=1
    cap.release()
    
    # Read images from viddir and write to imgdir.
    all_img_paths = sorted([os.path.join(viddir, ii) for ii in os.listdir(viddir)])
    for img_path in all_img_paths:
        img_ = cv2.imread(img_path)
        export_path = os.path.join(imgdir, os.path.basename(img_path))
        try:
            markers = ps3.find_markers(img_, tmpl)
            img2_ = ps3.draw_box(img_, markers, 3)
            dev_null = cv2.imwrite(export_path, img2_)
        except:
            print("Error processing {}".format(img_path))
            continue
            
    print("Finished {}.".format( print(vidpath)))
    
    

```
